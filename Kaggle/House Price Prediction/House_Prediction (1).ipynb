```python
from sklearn.preprocessing import LabelEncoder
from scipy.stats import zscore


col=df1.columns
le=LabelEncoder()
for i in col:
    df1[i]=le.fit_transform(df1[i])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report
from sklearn.ensemble import AdaBoostClassifier


X=df1.drop('SalePrice',axis=1)
y=df1['SalePrice']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
print(" X_train dataset: ", X_train.shape)
print(" y_train dataset: ", y_train.shape)
print(" X_test dataset: ", X_test.shape)
print(" y_test dataset: ", y_test.shape)
```

```python
from scipy.stats import zscore
XScaled=X.apply(zscore)
```

```python
tree = DecisionTreeRegressor()
tree.fit(X_train, y_train)
y_pred_tree = tree.predict(X_test)

print('Accuracy on Training data:',tree.score(X_train, y_train) )  
print('Accuracy on Test data:',tree.score(X_test, y_test) )
```

```python
pca1=PCA(n_components=60)
pca1.fit(XScaled)
```

```python
x=Xpca
y=df['SalePrice']


X_train_pca, X_test_pca, y_train_pca, y_test_pca = train_test_split(x,y, test_size = 0.3, random_state=10)

print(" X_train-pca dataset: ", X_train_pca.shape)
print(" y_train_pca dataset: ", y_train_pca.shape)
print(" X_test_pca dataset: ", X_test_pca.shape)
print(" y_test_pca dataset: ", y_test_pca.shape)
```

```python
tree_pca = DecisionTreeRegressor()
tree_pca.fit(X_train_pca, y_train_pca)
y_pred_tree_pca = tree_pca.predict(X_test_pca)

print('Accuracy on Training data:',tree_pca.score(X_train_pca, y_train_pca) )  
print('Accuracy on Test data:',tree_pca.score(X_test_pca, y_test_pca) )
```
